# 📊 Feature Importance Analysis for Stock Forecasting

| **Feature**           | **Type**                | **Short-Term (2-3 months) 📈 | **Long-Term (12-18 months) 📊 | **Reasoning** |
|----------------------|----------------------|----------------|----------------|----------------------|
| **Date**            | Time-Based           | ✅ Yes (Seasonality) | ✅ Yes (Trend) | Important for trend analysis and economic cycles |
| **Close**           | Price Data           | ✅ Yes  | ✅ Yes  | Primary target variable for stock prediction |
| **Volume**          | Market Data          | ✅ Yes  | ❌ No   | Affects short-term volatility but less useful long-term |
| **Dividends**       | Financial Metric     | ❌ No   | ✅ Yes  | Affects investor confidence and long-term valuation |
| **Stock Splits**    | Corporate Events     | ❌ No   | ✅ Yes  | Stock splits impact valuation and historical prices |
| **1-Day Price Change (%)** | Momentum Indicator | ✅ Yes  | ❌ No | Useful for short-term price momentum but noisy long-term |
| **SMA_50 (50-day Moving Avg)** | Technical Indicator | ✅ Yes | ✅ Yes  | Short-term momentum and long-term trend indicator |
| **SMA_200 (200-day Moving Avg)** | Technical Indicator | ❌ No  | ✅ Yes | Strong signal for long-term trend reversal |
| **RSI (Relative Strength Index)** | Momentum Indicator | ✅ Yes  | ❌ No | Identifies short-term overbought/oversold conditions |
| **MACD (Moving Average Convergence Divergence)** | Trend Indicator | ✅ Yes  | ✅ Yes | Signals trend direction and momentum shifts |
| **Volatility** | Risk Indicator | ✅ Yes  | ✅ Yes | High volatility affects short-term risk, stable trends help long-term forecasts |
| **Day of Week** | Categorical | ✅ Yes  | ❌ No | Market patterns exist (e.g., Monday effect, Friday effect) |
| **Inflation** | Macro Indicator | ❌ No | ✅ Yes | Inflation impacts interest rates and long-term stock trends |
| **GDP Growth** | Macro Indicator | ❌ No | ✅ Yes | Strong GDP growth boosts investor confidence over time |
| **Interest Rate** | Macro Indicator | ✅ Yes  | ✅ Yes | Affects bond yields and stock market valuation |
| **Yield Spread (10Y - 2Y Bond Yield)** | Recession Indicator | ❌ No | ✅ Yes | Yield curve inversion predicts recessions, affecting stocks long-term |

```python
import numpy as np
# Drop NaNs if any
series = mcd_data['Close']

# Apply FFT
fft_vals = np.fft.fft(series)
fft_freqs = np.fft.fftfreq(len(series))

# Keep only the positive frequencies
positive_freqs = fft_freqs[fft_freqs > 0]
positive_fft_vals = np.abs(fft_vals[fft_freqs > 0])

# Plot
plt.figure(figsize=(15,5))
plt.plot(positive_freqs, positive_fft_vals)
plt.title('FFT Spectrum - Dominant Frequencies')
plt.xlabel('Frequency')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

```python
# Get index of max amplitude
dominant_idx = np.argmax(positive_fft_vals)
dominant_freq = positive_freqs[dominant_idx]

# Convert to period (in days)
dominant_period = 1 / dominant_freq
print(f"Dominant period: {dominant_period:.2f} days")
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

endog = mcd_data['Close']

exog_vars = ['Dividends', 'Stock Splits','MACD', 'volatility', 'Inflation', 'GDP_Growth',
             'Interest_Rate', 'Yeild spread']# 'SMA_50', 'SMA_200'

exog = mcd_data[exog_vars]

# scaler = StandardScaler()
# exog_scaled = scaler.fit_transform(exog)
```

```python
order = (3, 1, 2)
seasonal_order = (3,1,1,52)

model =SARIMAX(endog, exog=exog,
                       order=order,
                       seasonal_order=seasonal_order,
                       enforce_stationarity=False,
                       enforce_invertibility=False)

results = model.fit(disp=False)
```

```python
print(results.summary())
```

```python
forecast_steps=180

rolling_window_exog = exog.iloc[-60:].values

future_exog = np.tile(rolling_window_exog, (forecast_steps // 60, 1))
fe=pd.DataFrame(future_exog)
fe.head()
```

```python

forecast = results.get_forecast(steps=forecast_steps, exog=future_exog)


forecast_mean = forecast.predicted_mean
conf_int = forecast.conf_int()

forecast_lower = conf_int.iloc[:, 0]
forecast_upper = conf_int.iloc[:, 1]

forecast_df = pd.DataFrame({
    'Date': pd.date_range(start=mcd_data.index[-1] + pd.Timedelta(days=1), periods=forecast_steps),
    'Forecast_Close': forecast_mean,
    'Lower_CI': forecast_lower,
    'Upper_CI': forecast_upper
})

print(forecast_df.head())
```

```python
plt.figure(figsize=(12,6))
plt.plot(mcd_data.index, endog, label='Observed')
plt.plot(forecast_df['Date'], forecast_df['Forecast_Close'], color='orange', label='Forecast')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.title('MCD Stock Price Forecast (SARIMAX)')
plt.legend()
plt.grid(True)
plt.show()
```
